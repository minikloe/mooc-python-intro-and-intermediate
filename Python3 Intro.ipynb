### 4-7 Python之continue继续循环  
>使用continue，我们可以控制循环继续下去，并跳过continue后面的逻辑，比如，对于字符串s = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'，假如希望输出字符串s中第10个以后的字符，而不是所有字符，这个时候， 我们可以使用continue跳过前面的9个字符。 

```python
s = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
num = 1
for ch in s:
    if num < 10:
        num = num + 1
        continue # 当num < 10时，跳过后续循环代码，继续下一次循环
    print(ch)
    num = num + 1
```
>*my understanding*：continue跳过continue 所对应的计算，不包含本次计算结果，但是继续到下一次计算中，下一次计算如果num还是<10，那么同样不包含这个结果，一直continue到num>10的时候，就开始正常运算了

### 6-1 什么是tuple 
 >元组(tuple)和list一样，也是一个有序容器，\
 在元组中，同样可以包含0个或者多个元素，也**支持索引访问、切片等操作**。\
定义元组的方式是使用小括号()将元组内的元素括起来。\

>元组数据类型可以**把不是元组的容器转换为元组**，比如将列表转换成元组。
```
T = tuple(L)
```
>tuple和list不一样的是，**tuple是固定不变的，一旦变成tuple，tuple中的每一个元素都不可被改变**，同时也不能再往tuple中添加数据，而list是可以的\
Python list method list() takes sequence types and converts them to lists. This is used to convert a given tuple into list.\
Note − Tuple are very similar to lists with only difference that element values of a tuple can not be changed 


>元组(tuple)的这个特性是非常重要的，在运行上tuple的性能是list的数倍.在实际编程中，tuple经常用于存放固定不变的数据。


### 6-2 访问tuple元素的其他方法  
>count()方法用来统计tuple中某个元素出现的次数。  
```python
T = (1, 1, 2, 2, 3, 3, 1, 3, 5, 7, 9)
print(T.count(1)) # ==> 3
```
count方法可用于: string, number, list, tuple,

>index()方法可以返回指定元素的下标，当一个元素多次重复出现时，则返回第一次出现的下标位置。


### 6-3 Python创建单个元素的tuple  
 >要定义只有一个元素的tuple，需要在元素后面添加一个逗号`,`
 为了与运算是表达优先级 同样用的`（ ）`区分开  
 如果这么写：   
```python 
T = (1)
print(T) # ==> 1
```
python会只给出`1`，而不是带括号的`(1)`，`(1)`才是元组，只有`1`代表python解释器take it as 计算结果了。  
因此，要定义只有一个元素的tuple，需要在元素后面添加一个逗号`,`  

>但对于多个元素的tuple，则加和不加这个逗号，效果是一样的。
```python
>>> T = (1, 2, 3,)
>>> print(T) # ==> (1, 2, 3)
```

>tuple的元素也可以是tuple
```python
T = ((1+2),  ((1+2),), ('a'+'b'), (1, ), (1,2,3,4,5))
```
>第一个元素(1+2)，第三个元素('a'+'b')，计算的结果分别是数字3和字符串ab，
 
### 6-4 Python的可变tuple  
 >说tuple不可变，仅针对：仅包含基础数据类型（数字类型、布尔类型、字符串类型）的数据，对于组合数据类型，则不受这个约束。
 ```python
T = (1, 'CH', [3, 4])
```
这里T有三个元素，第一个元素是数字类型，第二个元素是字符串类型，第三个元素是列表类型的，我们尝试修改第三个元素的数据:

```python
T = (1, 'CH', [3, 4])
L = T[2]
print(L) # ==> [3, 4]
# 尝试替换L中的元素
L[1] = 40
print(L) # ==> [3, 40]
print(T) # ==> (1, 'CH', [3, 40])
```
我们发现，元组T中的第三个元素已经改了，这有悖前面定义：元组是不可改变的。为什么呢？
因为虽然tuple中的list元素改变了，但是tuple本身指向的list仍然是同一个list，list本身并没有改变，改变的只是list里面的一个元素，这是tuple所约束不到的范围。  
但如果直接替换list，就不行。

---
2. 通过dict提供的items()方法，items()方法会返回dict中所有的元素，每个元素包含key和value。


```python
for key, value in d.items():
    if value > 60:
        print(key, value)
# ==> Candy 75
# ==> David 86
```

### 8-1 什么是set  
>**用途**： 有时，只想要key，不管value，只想要集合的元素不重复，就用`set`    
**组成**：set类似list，有一系列元素，
**区别**：    1. 元素不重复（list可有相同元素）
             2. 元素无序
**创建set**：用set()，传入一个list，list的元素将会被转换成set的元素。

```python
s = set([1, 1，4, 3, 2, 5])
print(s) # ==> set([1, 2, 3, 4, 5])
```
注意： 
1. 结果类似 list， 但不是 list  
2. 结果顺序与原来不同，因为 set无序
3. 如传入相同元素，set只保留一个

### 9-2 Python定义函数

依次写出 函数名、括号()、括号中的参数和冒号:  
然后，在缩进块中编写函数体，

>函数内部通过**条件判断和循环**可以实现非常复杂的逻辑。

### 9-3 Python函数返回值 
1. 返回None  
```python
def list_sum(l):
    result = 0
    for num in l:
        result = result + num
    print('result is {}'.format(result))
    return
```
在`print(result)`中，我们得到None的结果，这是合理的，
因为在函数内部，我们把结果打印出来了，但是没有把结果返回



2. 返回多个值  
除返回**None、一个值**外，函数也可返回多值.  
在函数中，如需要返回多个值，多值用逗号`,`分，注意顺序    
ex）定义函数data_of_square，接收边长一个参数，返回正方形周长、面积
```python
def data_of_square(side):
    C = 4 * side
    S = side * side
    return C, S
C, S = data_of_square(16)
print('周长 = {}'.format(C)) # ==> 周长 = 64
print('面积 = {}'.format(S)) # ==> 面积 = 256
```
3. 也可一个值 储函数返回的多值结果
```python
result = data_of_square(16)
print(result) # ==> (64, 256)
```
注：打印的result，是tuple类型。
如需取出结果中的周长、面积，使用对应位置下标 获得对应的结果
```python
result = data_of_square(16)
C = result[0]
S = result[1]
print('周长 = {}'.format(C)) # ==> 周长 = 64
print('面积 = {}'.format(S)) # ==> 面积 = 256
```

### 9-4 Python递归函数  
>函数内部，还可以调用其他函数. 
如果在一个函数内部调用其自身，这个函数就是递归函数。  
理论上，**所有的递归函数都可以写成循环**的方式，但循环的逻辑不如递归清晰  

>使用递归函数需要注意防止栈溢出。  
在计算机中，函数调用是通过栈（stack）这种数据结构实现的，
每当进入一个函数调用，栈就会加一层栈帧，
每当函数返回，栈就会减一层栈帧。
由于栈的大小不是无限的，所以，递归调用的次数过多，会导致栈溢出。

### 9-7 Python函数使用可变参数  `*args`

可变参数即任意个参数的意思，可变参数通常使用`*args`来表示: 
```python
def func(*args):
```

在使用上，Python会把可变参数定义为一个 tuple，  
∴ 在函数内部，把可变参数当作tuple来使用就可以了，如：可通过位置下标取出对应的元素等。

**缺点**：tuple在使用上有一定的局限性  
比如有时候想找到特定位置的参数，只能通过下标的方式去寻找  
如果顺序发生变化得时候，下标就会失效，函数逻辑就得重新修改实现  


定义可变参数的目:简化调用。
假设我们要计算任意个数的平均值，就可以定义一个可变参数：
```python
def average(*args):
    sum = 0
    for item in args:
        sum += item
    avg = sum / len(args)
    return avg
```


### 9-8 Python函数使用 可变关键字参数 `**kwargs`

**可变参数·缺点**：tuple在使用上有一定的局限性  
比如有时候想找到特定位置的参数，只能通过下标的方式去寻找  
如果顺序发生变化得时候，下标就会失效，函数逻辑就得重新修改实现 

**可变关键字参数**：可以通过关键字的名字key找到对应的参数值，  
这和dict相似，Python会把可变关键字参数当作dict去处理；  
对于可变关键字参数，一般使用`**kwargs`来表示。  

**各种参数顺序**
对于一个拥有：必需参数，默认参数，可变参数，可变关键字参数 的函数，
定义顺序是这样的：

def func(param1, param2, param3 = None, *args, **kwargs):
    print(param1)
    print(param2)
    print(param3)
    print(args)
    print(kwargs)
```python
func(100, 200, 300, 400, 500, name = 'Alice', score = 100)
# ==> 100
# ==> 200
# ==> 300
# ==> (400, 500)
# ==> {'name': 'Alice', 'score': 100}
```

当然，这么多类型的参数，容易导致出错，在实际使用上，不建议定义这么多的参数。
